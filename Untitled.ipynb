```python
import yfinance as yf
import pandas as pd

# Amazon şirketinin verilerini Yahoo Finance üzerinden çekiyoruz
df = yf.download("AMZN", start="2021-01-01", end="2026-01-01")

# İndirdiğimiz verinin ilk 5 gününe göz atalım
df.head()
```

```python
import matplotlib.pyplot as plt

# Sadece kapanış (Close) fiyatlarını çizdireceğiz
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Amazon Kapanış Fiyatı', color='blue')
plt.title('Amazon Hisse Senedi Geçmişi (2021-2026)')
plt.xlabel('Tarih')
plt.ylabel('Dolar ($)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import torch
from sklearn.preprocessing import MinMaxScaler

# 1. Sadece Kapanış fiyatlarını alıp yapay zekaya uygun formata getiriyoruz
close_prices = df['Close'].values.reshape(-1, 1)

# 2. Verileri -1 ile 1 arasına sıkıştırıyoruz (Normalizasyon)
scaler = MinMaxScaler(feature_range=(-1, 1))
scaled_data = scaler.fit_transform(close_prices)

# 3. Kayan Pencere (Sliding Window) fonksiyonu: Geçmiş verilerden parça koparma
def create_sequences(data, lookback=20):
    X = [] # Giriş: Son 20 günün fiyatları
    y = [] # Hedef: 21. günün fiyatı
    for i in range(len(data) - lookback):
        X.append(data[i:(i + lookback)])
        y.append(data[i + lookback])
    return np.array(X), np.array(y)

# Son 20 günü baz alarak verileri paketliyoruz
lookback_window = 20
X, y = create_sequences(scaled_data, lookback_window)

# 4. Verileri PyTorch'un anlayacağı "Tensor" formatına çeviriyoruz
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)

print("Giriş Verisi (X) Şekli:", X_tensor.shape)
print("Hedef Verisi (y) Şekli:", y_tensor.shape)
```

```python
import torch.nn as nn

# 1. LSTM Model Tasarımı
class LSTMModel(nn.Module):
    def __init__(self, input_dim=1, hidden_dim=32, num_layers=2, output_dim=1):
        super(LSTMModel, self).__init__()
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        # LSTM Katmanı: Geçmiş bilgileri uzun vadeli hafızada tutar
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True)
        # Doğrusal Katman: Çıkışı tek bir tahmin fiyatına dönüştürür
        self.fc = nn.Linear(hidden_dim, output_dim)
        
    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim)
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :]) # Sadece son günün çıktısını alıyoruz
        return out

# 2. GRU Model Tasarımı
class GRUModel(nn.Module):
    def __init__(self, input_dim=1, hidden_dim=32, num_layers=2, output_dim=1):
        super(GRUModel, self).__init__()
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        # GRU Katmanı: LSTM'e benzer ama daha az parametrelidir, daha hızlı çalışır
        self.gru = nn.GRU(input_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
        
    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim)
        out, _ = self.gru(x, h0)
        out = self.fc(out[:, -1, :]) # Sadece son günün çıktısını alıyoruz
        return out

# Modelleri oluşturup test edelim
lstm_model = LSTMModel()
gru_model = GRUModel()

print("LSTM Modeli Başarıyla Kuruldu!")
print("GRU Modeli Başarıyla Kuruldu!")
```

```python
import torch.optim as optim
import time

# 1. Veriyi Eğitim ve Test olarak ikiye bölüyoruz (%80 Eğitim, %20 Test)
train_size = int(len(X_tensor) * 0.8)
X_train, X_test = X_tensor[:train_size], X_tensor[train_size:]
y_train, y_test = y_tensor[:train_size], y_tensor[train_size:]

# 2. Eğitim Ayarları (Hata fonksiyonu ve Güncelleyici)
criterion = nn.MSELoss() # Regresyon için Ortalama Kare Hata fonksiyonu
epochs = 50

# --- LSTM EĞİTİMİ ---
print("--- LSTM Modeli Eğitiliyor ---")
lstm_optimizer = optim.Adam(lstm_model.parameters(), lr=0.01)
start_time = time.time()

for epoch in range(epochs):
    lstm_model.train()
    lstm_optimizer.zero_grad()
    
    outputs = lstm_model(X_train)
    loss = criterion(outputs, y_train)
    
    loss.backward()
    lstm_optimizer.step()
    
    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Hata (Loss): {loss.item():.4f}")

lstm_time = time.time() - start_time
print(f"LSTM Eğitimi Bitti! Süre: {lstm_time:.2f} saniye.\n")


# --- GRU EĞİTİMİ ---
print("--- GRU Modeli Eğitiliyor ---")
gru_optimizer = optim.Adam(gru_model.parameters(), lr=0.01)
start_time = time.time()

for epoch in range(epochs):
    gru_model.train()
    gru_optimizer.zero_grad()
    
    outputs = gru_model(X_train)
    loss = criterion(outputs, y_train)
    
    loss.backward()
    gru_optimizer.step()
    
    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Hata (Loss): {loss.item():.4f}")

gru_time = time.time() - start_time
print(f"GRU Eğitimi Bitti! Süre: {gru_time:.2f} saniye.")
```

```python
# 1. Modelleri test moduna alıp tahmin üretiyoruz
lstm_model.eval()
gru_model.eval()

with torch.no_grad():
    lstm_predictions = lstm_model(X_test).numpy()
    gru_predictions = gru_model(X_test).numpy()

# 2. Sıkıştırılmış verileri (-1, 1 aralığından) tekrar gerçek Dolar fiyatlarına döndürüyoruz
lstm_predictions_real = scaler.inverse_transform(lstm_predictions)
gru_predictions_real = scaler.inverse_transform(gru_predictions)
y_test_real = scaler.inverse_transform(y_test.numpy())

# 3. Sonuçları Karşılaştırmalı Grafik Olarak Çizdiriyoruz
plt.figure(figsize=(14, 7))
plt.plot(y_test_real, label='Gerçek Amazon Fiyatı', color='black', linewidth=2)
plt.plot(lstm_predictions_real, label='LSTM Tahmini', color='blue', linestyle='--')
plt.plot(gru_predictions_real, label='GRU Tahmini', color='green', linestyle='-.')

plt.title('Yapay Zeka Borsa Tahmin Performansı (Test Verisi)')
plt.xlabel('Zaman (Test Günleri)')
plt.ylabel('Fiyat ($)')
plt.legend()
plt.grid(True)
plt.show()

# 4. Sayısal Hata Değerlerini Hesaplayıp Yazdıralım (MSE)
from sklearn.metrics import mean_squared_error
print(f"LSTM Test MSE (Hata Değeri): {mean_squared_error(y_test_real, lstm_predictions_real):.4f}")
print(f"GRU Test MSE (Hata Değeri): {mean_squared_error(y_test_real, gru_predictions_real):.4f}")
```
